# **Data Visualisation**

```python
# Visualize the Temprature Variations for March

import matplotlib.pyplot  as plt

# Convert 'Date/Time' feature to datetime format

data['Date/Time'] = pd.to_datetime(data['Date/Time'], format='%m/%d/%Y %H:%M')

# Determine rows where the month indicates March

march_data = data[data['Date/Time'].dt.month == 3]


average_temperature = []
average_pressure = []

# Determine the mean temperature and pressure for each day of March

for i in range(1,32) :

     march_data_day1 = march_data[march_data['Date/Time'].dt.day == i]['Temp_C'].mean()
     average_temperature.append(march_data_day1)
     
     march_data_day2 = march_data[march_data['Date/Time'].dt.day == i]['Press_kPa'].mean()
     average_pressure.append(march_data_day2)     

# Create subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(18, 5))

# Plotting on the first subplot
axes[0].plot(range(1, 32), average_temperature,'r')
axes[0].set_xlabel('Day')
axes[0].set_ylabel('Average Temperature (°C)')
axes[0].set_title('Temperature Variations of March')
axes[0].set_xticks(range(1, 32, 2))
axes[0].grid(True)


# Plotting on the second subplot
axes[1].plot(range(1, 32), average_pressure)
axes[1].set_xlabel('Day')
axes[1].set_ylabel('Average Pressure')
axes[1].set_title('Pressure Variations of March')
axes[1].set_xticks(range(1, 32, 2))
axes[1].grid(True)

# Adjust layout for better spacing
plt.tight_layout()

plt.show()     
```

```python
# display the visibility of the first week of march


average_wind_speed_march = []

april_data = data[data['Date/Time'].dt.month == 4]
average_wind_speed_april = []


days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']


for i in range(1,8) :

     march_data_week = march_data[march_data['Date/Time'].dt.day == i]['Wind Speed_km/h'].mean()
     average_wind_speed_march.append(march_data_week)
     
     april_data_week = april_data[april_data['Date/Time'].dt.day == i]['Wind Speed_km/h'].mean()
     average_wind_speed_april.append(april_data_week)
 
        

# Plotting

bar_width = 0.35  # Adjust the width of the bars

plt.figure(figsize=(13, 5))

plt.bar(range(1, 8), average_wind_speed_march, width=bar_width, color='cyan', label='Wind Speed of March')
plt.bar([day + bar_width for day in range(1, 8)], average_wind_speed_april, width=bar_width, color='pink', label='wind speed of April ')

plt.ylabel('Wind speed (km/h) ')
plt.title('Comparing Wind Speeds in the First Weeks of March and April')

# Set x-axis ticks and labels
plt.xticks([day + bar_width/2 for day in range(1, 8)], days)

plt.legend()
plt.show()

```
